# Data Visualization

```python
sns.pairplot(dataset)
```

```python
#Correlation between columns and visualising using heatmaps
correlation=dataset.corr()
sns.heatmap(correlation,annot=True)
plt.figure(figsize=(10,5))
plt.show()
```

```python
#Based on Category
sns.pairplot(dataset,hue="Category")
```

```python
#Sales vs Profit
plt.figure(figsize=(12,5))
plt.scatter(dataset['Sales'],dataset['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales VS Profit')
```

```python
#Sales vs Discount
plt.figure(figsize=(12,5))
plt.scatter(dataset['Sales'],dataset['Discount'])
plt.xlabel('Sales')
plt.ylabel('Discount')
plt.title('Sales VS Discount')
```

```python
#Countplot category wise
plt.figure(figsize=(8,5))
sns.countplot(dataset['Category'])
```

```python
#Countplot Sub-category wise
plt.figure(figsize=(17,5))
sns.countplot(dataset['Sub-Category'])
```

```python
#Region wise profit
plt.figure(figsize=(8,5))
sns.barplot(x=dataset['Region'],y=dataset['Profit'])
```

```python
#Region wise Discount
plt.figure(figsize=(8,5))
sns.barplot(x=dataset['Region'],y=dataset['Discount'])
```

```python
fig, pos = plt.subplots(2,2, figsize = (18, 10))
fig.suptitle("Sales with different shipping modes and Segments ", fontsize = 16)
sns.barplot(dataset['Ship Mode'], dataset['Sales'], ax=pos[0,0],dodge=False)
sns.lineplot(dataset['Ship Mode'], dataset['Sales'], ax=pos[0,1])
sns.barplot(dataset['Segment'], dataset['Sales'], ax=pos[1,0])
sns.lineplot(dataset['Segment'], dataset['Sales'], ax=pos[1,1])
plt.show()
```

```python

#Distribution using distplot
fig, pos = plt.subplots(2,2, figsize = (16, 8))
fig.suptitle("Distribution Plots", fontsize = 18)
sns.distplot(dataset['Sales'], ax=pos[0,0])
sns.distplot(dataset['Profit'], ax=pos[0,1])
sns.distplot(dataset['Discount'], ax=pos[1,0])
sns.distplot(dataset['Quantity'], ax=pos[1,1])
plt.show()
```

```python
#Profit VS Discount
plt.figure(figsize=(10,8))
plt.title('Profit VS Discount',fontsize=20)
sns.lineplot(dataset['Discount'], dataset['Profit'])
```

```python
#Sales VS category
plt.figure(figsize=(8,5))
plt.title('Sales VS Category',fontsize=16)
sns.barplot(x=dataset['Sales'],y=dataset['Category'])
```

```python
#Sales VS Sub-category
plt.figure(figsize=(16,10))
plt.title('Sales VS Sub-Category',fontsize=16)
sns.barplot(x=dataset['Sales'],y=dataset['Sub-Category'])
```
